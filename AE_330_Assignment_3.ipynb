# Rocket Stage Information

## Vikas Engine (PSLV Stage 2)

The second stage of the PSLV rocket is 2.8 meters in diameter, 12.8 meters long and has a liftoff mass of 46,000 Kilograms with an empty mass of 5,300 Kilograms.

The stage uses Unsymmetrical Di-Methylhydrazine / Hydrazine Hydrate as fuel and Nitrogen Tetroxide as oxidizer. It is powered by a single 799-Kilonewton Vikas engine. Vikas was developed by ISRO in the 1970s.

The engine has a dry weight of 900 Kilograms and operates at a chamber pressure of 58.5 bar and provides a specific impulse of 294 sec.

The second stage burns for 158 seconds. Vehicle control during second stage flight is accomplished by gimbaling the main engine by up to 4 degrees, roll control is provided by a Hot Gas Reaction Control Motor.

Second stage separation is accomplished with a merman band system and separation motors. It occurs at an altitude of 277 Kilometers.

## Cryogenic Upper Stage C-25 (GSLV Mk III Stage 2)

Starting with the second mission of GSLV Mk. III, the launcher will feature a Cryogenic Upper Stage carrying up to 28,000 Kilograms of Liquid Oxygen and Liquid Hydrogen. This upper stage is a larger version of the third stage of the Mk. II using a more powerful version of the Indian Cryogenic Engine. The upper stage measures about 13.55 meters in length and 4 meters in diameter with an inert mass on the order of five metric tons.

The third stage of the GSLV Mk. III is powered by a CE-20 cryogenic engine that is based on the CE-7.5 of the GSLV Mk. II, but delivers a much higher thrust of 200 Kilonewtons.

The engine is an open-cycle Gas Generator engine which uses a small fraction of the fuel and oxidizer flow from the turbopumps to burn it in a Gas Generator that delivers the high-pressure gas needed to drive the turbopumps of the fuel and oxidizer. The propellants are then injected into the combustion chamber where they are burned to generate thrust while the gas generator exhaust is ejected through a nozzle and not burned in the chamber.

The CE-20 engine operates at a fixed thrust level that can be set to 180 to 220 Kilonewtons. It operates at a mixture ratio of 5.05 and a nominal chamber pressure of 60 bar to achieve a specific impulse of 443 seconds.

The engine has a dry mass of about 588 Kilograms and uses an extended nozzle with an area ratio of 100 for optimized operation in vacuum. The engine provides re-start capability and operates up to 640 seconds in a nominal GSLV Mk. III flight.

Vehicle control during burns is provided by two vernier engines that can be swiveled in all directions to provide three-axis control. During Coast Phases, a cold gas reaction control system is used for vehicle stabilization and re-orientations. The spacecraft is separated by a tension release device mounted at the separation interface with the third stage.

The third stage of the launch vehicle also houses the flight computers and the inertial guidance platform of the GSLV. The control system was developed and built in India. GSLV uses a Redundant Strap Down Inertial Navigation System/Inertial Guidance System that is housed in the equipment bay of the third stage. The digital control system of the launcher uses closed-loop guidance throughout the flight to ensure accurate injections into the target orbit.

Also mounted on the third stage is the communications system of the launch vehicle consisting of an S-Band system for telemetry downlink and a C-Band transponder that allows radar tracking and preliminary orbit determination.

The communications link is also used for range safety / flight termination that uses a dedicated system that is located on all stages of the vehicle and features separate avionics.

## S200 Solid Roket Booster (GSLV Mk III Strapon boosters)

GSLV Mk. III uses two S200 solid-fueled boosters that are attached to the Core Stage and deliver thrust for the initial portion of the flight.

Also known as LSB (Large Solid Booster), each of the S200 boosters stands 26.2 meters tall and consists of three segments holding a total of 206,690 Kilograms of propellant.

The booster casing is 3.2 meters in diameter while the Flex Nozzle at the base of the booster is 3.27 meters in diameter.

S200 is the third largest Solid Rocket Booster used in any launch vehicle after the Space Shuttle and the Ariane 5 rocket. Like these two other vehicles, S200 uses a HTPB-based propellant (Hydroxyl-terminated Ploy-butadiene).

Major components of the S200 Solid Rocket Booster are the Flex Nozzle, the motor case, the propellant, igniter and nose cone.

The booster itself consists of three steel segments – the head-end segment contains 27,100kg of propellant, the middle segment carries 97,380kg and the Nozzle-end segment is loaded with 82,210kg of propellants.

S200 is ignited by a highly reliable pyrogen igniter that is initiated by a secondary add on pyrogen igniter which generates enough hot particles to travel down the length of the booster to ignite the propellant.

The igniter is based on that used on the S139 first stage of the PSLV and GSLV Mk. II launch vehicles.

It is triggered through a redundant safe and arm device that initiates explosive transfer assemblies.

The Flex nozzle at the base of the booster is divergent to direct the gas flow from the chamber.

S200 uses a nozzle with a throat diameter of 88.6 centimeters and an area ratio of 12.1. The Flex Nozzle assembly has a length of 3.58 meters consisting of a convergent subassembly, a flexseat subassembly, a submerged subassembly, a divergent fore-end subassembly and a divergent aft-end subassembly.

For vehicle control, the Flex Nozzle can be vectored using two electro-hydraulic servo actuators. This planar movement provides pitch and yaw control and, in combination with the other S200 booster, can be used for roll control as well.

The actuation system is housed in a flared base shroud installed on the booster that also builds the interface to the Mobile Launch Platform. The boosters use a dual redundant control unit with FDI logic.

The S200 boosters are designed to operate at a fixed pressure and thrust profile (M-type) with a maximum thrust of 5,151 Kilonewtons (sea level) and a average SL thrust of 3,578kN corresponding to chamber pressures of 56.9 and 39.9 bar.

S200 operates for 130 seconds and is separated from the launch vehicle at T+149 seconds using pyrotechnic separation devices and six small solid-fueled jettison motors located in the nose and aft segments of the boosters.

# Approximate Staging Solution

The required Δv for an LEO orbit of 400km is found to be 9400 m/s, factoring in the losses incurred due to drag and gravity. The launch is assumed to be from SDSC-SHAR with a latitude of θ = 13.7259°N.
For best result and easterly launch is preferred due to the effect of the Earth's rotation reducing the Δv.
Velcity imparted due to the Earth's rotation:

v$_{surf}$ = $\frac{2 π r_{e} cos(θ)}{1day}$ = 450.57 m/s

Thus required Δv = 9400-450.57 = 8949.42 m/s
Based on the velocity requirements, a two stage rocket is chosen with booster stages operating in parallel with the first stage during liftoff. This is because using a two stage rocket appears to have the optimal liftoff mass for the given mission requirements for the available propulsion technology. Addition of more stages causes small improvement in payload capacity while the complexity and weight increase is significant espaecially due to inert component weight.

The engines are chosen as follows :

Stage 0 - Two S200 strapons that operate in tandem with Stage 1 till their separation. $I_{sp,0}$ = 227s. Existing configuration of S200 will be used with adjustment to propellant mass and burn time accordingly.

Justification: The stage is chosen as such due to the requirement of high thrust during liftoff. Thus, the engines with higher thrust rating is chosen. A parallel configuration is chosen as it leads to lesser gravity losses for the upper stages as the strapons are discareded upon burnout.

Stage 1 - Vikas LRE stage with $I_{sp,1}$ = 330s (modified for higher $I_{sp}$) to continue operation till burnout after booster jettison. Existing configuration of Vikas engine will be used with adjustment to propellant mass and burn time accordingly.

Justification: The Vikas stage is chosen as such due to the requirement of high thrust and high $I_{sp}$ for the climb phase. Thus, the engines with higher $I_{sp}$ than  is chosen. 

Stage 2 - C 25 CUS with $I_{sp, vac, 2}$ = 443s. Existing configuration of C 25 will be used with adjustment to propellant mass and burn time accordingly.

Justification: The C25 stage is chosen as such due to the requirement of high $I_{sp}$ for the terminal phase. to reduce weight. This leads to lower payload mass for lower stages thereby causing lesser gravity losses and lower liftoff weight.

Based on current technology, the Δv requirements is shared between the stages as follows:

First stage in parallel with two boosters Δv = 2700 m/s

First stage after booster separation Δv = 2700 m/s

Second stage after booster separation Δv = 3549.42 m/s

## Stage 2 solution

$m_{f2}$ = $m_{pl}$ + $m_{i2}$ = 15 + 5 = 20 tons

Using Tsiolkovsky equation, we get

3549.42 = - $I_{sp,vac,2} g_{LEO}ln(\frac{m_{f2}}{m_{o2}})$

Using $I_{sp,vac,2}$ = 443 and $g_{LEO}$ = 8.8 m/$s^{2}$, we get $\frac{m_{f2}}{m_{o2}}$ = 0.4023

Using $m_{f2}$ = 20 tons, we get $m_{o2}$ = 49.71042 tons

This gives $m_{p2}$ = 29.71 tons

## Stage 1 solution after booster separation

$m_{f1}$ = $m_{02}$ + $m_{i1}$ = 49.71 + 5.3 = 55.01 tons

Using Tsiolkovsky equation, we get

2700 = - $I_{sp,1} g_{e}ln(\frac{m_{f1}}{m_{o1}})$

Using $I_{sp,1}$ = 330 and $g_{e}$ = 9.8 m/$s^{2}$, we get $\frac{m_{f1}}{m_{o1}}$ = 0.4339

Using $m_{f1}$ = 55.01 tons, we get $m_{o1}$ = 126.773 tons

This gives $m_{p1}$ = 71.763 tons

## Stage 0 (2 boosters) and Stage 1 parallel operation solution 

$m_{f0}$ = $m_{01}$ + 2×$m_{i0}$ = 127.773 + 2×31.3 = 189.373 tons

Since here two different engines are oprating in parallel, the effective $I_{sp}$ can be computed as follows:

$I_{sp, eff}$ = $\frac{(2×I_{sp,0}×\dot{m}_{0} + I_{sp,1}×\dot{m}_{1})}{2×\dot{m}_{0} + \dot{m}_{1}}$ = 235.28 s

Using Tsiolkovsky equation, we get

2700 = - $I_{sp,eff} g_{e}ln(\frac{m_{f0}}{m_{o}})$

Using $I_{sp,1}$ = 330 and $g_{e}$ = 9.8 m/$s^{2}$, we get $\frac{m_{f1}}{m_{o1}}$ = 0.3100

Using $m_{f0}$ = 189.373 tons, we get $m_{o}$ = 611.61 tons

This gives $m_{p0}$ = 422.417 tons

But this mass is given the combined propellant mass of the two boosters and the portion of propellant consumed by stage 1 during the parallel oparation with the boosters.

Since the mass flow rate of both booster and first stage is available and assumimg constant burn profile, the burn time for parallel burn of boosters with stage 1 can be found as:

$t_{b,parallel}$ = $\frac{422.417×1000}{2×\dot{m}_{0} + \dot{m}_{1}}$ = 122.02 s

Using the burn time and the propellant flow rates of the two engines, we can calculate the propellant masses of the engines as shown:

Taking $\dot{m}_{0}$ = 1589.92 kg/s and $\dot{m}_{1}$ = 278.04 kg/s we get $m_{p, booster}$ = 194.195 tons each and $m_{p1, parallel}$ = 34.126 tons. 

Thus the total propellant mass of stage 1 is $m_{p1, total}$ = 34.126 + 71.763 = 105.889 tons

# Engine Parameters for Each Stage

## Vikas (Stage 1 Engine)

From earlier design, $m_{p2}$ = 105.889 tons and $\dot{m}$ = 278.04 kg/s

We get the burn time as follows

$t_{b}$ = $\frac{m_{p1}}{\dot{m}}$ = 380s

$I_{sp,1}$ = 330 s

Using $I_{sp}$ = $\frac{T}{\dot{m}g_{e}}$, we can compute the thrust 

T = 899.18 kN at sea level (using $g_{e}$ = 9.8 m/$s^{2}$)

Given that for $N_{2}O_{4}$ + UDMH propellant, the characteristic velocity 
c* = 1652 m/s and the chamber pressure is given to be $P_{c}$ 58.5 bar = 5.85 MPa, the Throat area $A_{t}$ is calculated as follows

c* = $\frac{P_{c}A_{t}}{\dot{m}}$

This gives $A_{t}$ = 0.0785 $m^{2}$

The effective exhaust velocity is calculated as follows using $I_{sp,1}$ and $g_{e}$

c = $I_{sp,1}$×$g_{e}$ 

which gives c = 3234 m/s

The chamber pressure is given to be $P_{c}$ 58.5 bar = 5.85 MPa

The thrust coefficient $c_{f}$ can thus be found out using Thrust, $P_{c}$ and $A_{t}$ as shown

$c_{f}$ = $\frac{T}{P_{c}A_{t}}$ = 1.958

# Verification of Δv and Acceleration Constraints

The values of accelerations for each stage can be computed as follows

a(t) = $\frac{T - (m_{0}-\dot{m}t)gcos(ωt)}{m_{0}-\dot{m}t}$

## Stage 2 

$m_{0,2}$ = 49.71 tons

$\dot{m}_{2}$ = 0.03977 tons/s

T = 155 kN
 
g = 8.8 m/$s^{2}$ 

$a_{2}$(t) = $\frac{155 -(49.71-0.03977t)×8.8×cos(ωt)}{49.71-0.03977t}$

Assuming vehicle attitude to be almost horizontal during terminal phase and thus taking ωt = constant = 90° we get 

$a_{2}$(t) = $\frac{155}{49.71-0.03977t}$

Plotting for t = 0 s to 800 s (burn duration), we see that the maximum attained acceleration is given as

$η_{max, 2}$ = 8.662 m/$s^{2}$ = 0.9843 g
